```python
# csv 파일 읽고 저장하기
import numpy as np
import pandas as pd
ar1 = np.arange(10).reshape(2,5)
df = pd.DataFrame(ar1,columns=list('abcde'))
df.to_csv('dataset/test.csv')
```

```python
df = pd.read_csv('dataset/test.csv',index_col=0)
df
```
